```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df.columns
set(df[df['parent_category_name']== 'Услуги']['category_name'])
```

```python
df[df['parent_category_name']== 'Личные вещи'].to_csv('personal_thing.csv', index = False)
```

```python
plt.figure(figsize=(8, 5))
count_by_parent_category = df.groupby('parent_category_name')['item_id'].count()
count_by_parent_category.sort_values(ascending = False, inplace = True)
count_by_parent_category.plot(kind = 'bar')
plt.title('Количество объявлений')
plt.xticks( rotation='vertical') 
```

```python
plt.figure(figsize=(8, 5))
count_by_parent_category = df.groupby('parent_category_name')['price'].sum()
count_by_parent_category.sort_values(ascending = False, inplace = True)
count_by_parent_category.plot(kind = 'bar')
plt.title('Оборот в рублях')
plt.xticks( rotation='vertical') 
```

```python
plt.figure(figsize=(8, 5))
pers_things = df[df['parent_category_name'] == 'Личные вещи'].groupby('category_name')['item_id'].count()
pers_things.sort_values(ascending = False, inplace = True)
pers_things.plot(kind = 'bar')
plt.title('Оборот в рублях')
plt.xticks( rotation='vertical') 
```

```python
plt.figure(figsize=(8, 5))
count_by_parent_category = df.groupby('user_type')['item_id'].count()
count_by_parent_category.sort_values(ascending = False, inplace = True)
count_by_parent_category.plot(kind = 'bar')
plt.title('Количество объявлений')
plt.xticks( rotation='vertical') 
```

```python
mean_prob_user_type = df.groupby('user_type')['deal_probability'].mean()
mean_prob_user_type.sort_values(ascending = False, inplace = True)
mean_prob_user_type.plot(kind = 'bar')
plt.title('Средняя вероятность')
plt.xticks( rotation='vertical') 
```

```python
plt.figure(figsize=(8, 10))
count_by_city = df.groupby('region')['item_id'].count()
count_by_city.sort_values(ascending = True, inplace = True)
count_by_city.plot(kind = 'barh')
plt.xticks( rotation='vertical')
```
