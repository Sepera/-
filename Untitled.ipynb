```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
from datetime import datetime, date, time
from geopy.distance import great_circle
```

```python
data = pd.read_csv('Data.csv', encoding='Windows-1251') 
```

```python
data['Дата и время'] = pd.to_datetime(data['Дата и время'], format='%d.%m.%Y %H:%M')
data['Дата'] = data['Дата и время'].map(lambda x: x.date())
```

```python
# Визуализируется это с помощью графика, где по оси абсцисс откладываются дни (т.е. сутки), 
# а по оси ординат - количество сигналов.
daily_activity = data.groupby(['Дата', 'Объект'])['Широта'].count().reset_index().sort_values('Дата')
daily_activity1 = pd.pivot_table(daily_activity, index='Дата', columns='Объект', values='Широта')
daily_activity1.plot(figsize=(12,8))
plt.show()
```

```python
#Должен быть сделан вывод в виде текста, какой объект более активен ночью, какой утром, какой днём, какой вечером.
daily_activity.merge(pd.pivot_table(daily_activity, index = ['Объект'],values='Широта', aggfunc=np.max).reset_index()).rename(columns = {'Широта': 'Активность'})
```

```python
#Визуализируется это с помощью графика, где по оси абсцисс откладываются часы 
# (24 значения), а по оси ординат - количество сигналов.
data['Время'] = data['Дата и время'].map(lambda x: x.hour)
day_activity = data.groupby(['Время', 'Объект'])['Широта'].count().reset_index().sort_values('Время')
pd.pivot_table(day_activity, index='Время', columns='Объект', values='Широта').plot(figsize=(12,8))
plt.show()
```

```python
day_activity = data.groupby(['Время', 'Объект'])['Широта'].count().reset_index()
pd.pivot_table(day_activity, index = ['Объект'],values='Широта', aggfunc=np.max).reset_index()
day_activity.merge(pd.pivot_table(day_activity, index = ['Объект'],values='Широта', aggfunc=np.max).reset_index()).rename(columns = {'Широта': 'Активность'})
```

```python
def day_time(x):
    if x < 6:
        return 'Ночь'
    elif x > 5 and x < 12:
        return 'Утро'
    elif x > 11 and x < 18:
        return 'День'
    else:
        return 'Вечер'
data['Время суток'] = data['Время'].map(lambda x: day_time(x))
```

```python
day_activity = data.groupby(['Время суток', 'Объект'])['Широта'].count().reset_index()
day_activity1 = pd.pivot_table(day_activity, index='Время суток', columns='Объект', values='Широта')
day_activity1 = pd.DataFrame(index = ['Ночь', 'Утро', 'День', 'Вечер'], data= day_activity1)
day_activity1.plot(figsize=(12,8), kind='bar')
plt.show()
```

```python
#Должен быть сделан вывод в виде текста, какой объект более активен ночью, какой утром, какой днём, какой вечером.
pd.pivot_table(day_activity, index = ['Объект'],values='Широта', aggfunc=np.max).reset_index()
day_activity.merge(pd.pivot_table(day_activity, index = ['Объект'],values='Широта', aggfunc=np.max).reset_index()).rename(columns = {'Широта': 'Активность'})
```

```python
'''
делает поиск ситуаций, когда объекты подходили достаточно близко друг к другу. 
Т.е. необходимо найти все пары сигналов от разных объектов, отличающиеся на время не большее T, 
а расстояние между ними должно быть не больше D.
Параметры T (в минутах) и D (в метрах) должны задаваться пользователем перед началом анализа данных.
'''
T = input("Введите временной интервал (минуты): ")
L = input("Введите простанственный интервал (метры): ")
for i in range(len(data) - 1):
    j = i + 1
    while j < len(data):
        k = great_circle((data['Широта'][i],data['Долгота'][i]), (data['Широта'][j], data['Долгота'][j])).meters
        time_i = data['Дата и время'][i]
        time_j = data['Дата и время'][j]
        if ((time_j < (time_i + pd.tseries.offsets.Minute(T))) and (k < int(L))) or (time_j < time_i + pd.tseries.offsets.Minute(T)):            
            if data['Объект'][i] != data['Объект'][j] and k < int(L):
                print('Расстояние(',data['Объект'][i], data['Объект'][j],') =', k)
                j = j +1
            else:
                j = j + 1
                continue
        else:
            break
```
